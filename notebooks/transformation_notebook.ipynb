```python
# 1. Define your data using a dictionary format (list of dicts)
data = [
    {"employee_id": 101, "name": "Alice", "department": "Engineering", "salary": 95000},
    {"employee_id": 102, "name": "Bob", "department": "Data Science", "salary": 105000},
    {"employee_id": 103, "name": "Charlie", "department": "Engineering", "salary": 88000},
    {"employee_id": 104, "name": "Diana", "department": "Product", "salary": 110000}
]

# 2. Create the Spark DataFrame
# Spark automatically infers the schema/column names from the dictionary keys
df = spark.createDataFrame(data)

# 3. Write the DataFrame to a Delta Table
# Replace 'default' with your specific database/schema name if needed
table_name = "default.employee_sample_data"

df.write \
  .mode("overwrite") \
  .saveAsTable(table_name)

print(f"Successfully wrote data to table: {table_name}")
```
